```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
lunch=df[df["time"] == "Lunch"]
Dinner=df[df["time"] == "Dinner"]
lunch.plot("total_bill","tip",kind="scatter",c="b")
Dinner.plot("total_bill","tip",kind="scatter",c="green")

```

```python
sns.violinplot(x = "sex", y = "tip", data = df, hue = "time")
```

```python
sns.catplot( x = "sex", y = "tip", hue = "time", data = df )
```

```python
sns.pairplot(data = df, hue = "smoker")
```

```python
sns.jointplot("total_bill", "tip", data=df, kind="hex", height = 10)
```

```python
sns.catplot(x="time", y="total_bill", hue="sex", col="sex", data = df)
```
